```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
hostels=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
hostels.send_keys("London")
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]')
search.click()
```

```python
hostel_url =[]
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="nuxt-link"]')
    for i in url:
        hostel_url.append(i.get_attribute('href'))
    time.sleep(2)    
```

```python
Hostel_name=[]
PD=[]
Rating=[]
Dis=[]

for url in hostel_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        name=driver.find_elements(By.XPATH,'//div[@class="headline-container"]')
        Hostel_name.append(name.text)
    except NoSuchElementException :
        Hostel_name.append('-')
              
     
    try:
        description=driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[2]/div[4]')
        PD.append(description.text)
    except NoSuchElementException :
        PD.append('-')
        
    try:
        ratings=driver.find_elements(By.XPATH,'//div[@class="score orange big"]')
        Rating.append(ratings.text)
    except NoSuchElementException:
        Rating.append('-')
        
    try:
        distance=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
        Dis.append(distance.text)
    except NoSuchElementException:
        Dis.append('-')
           
```
